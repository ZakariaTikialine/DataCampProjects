```python
# Import necessary packages
import pandas as pd
import numpy as np

# Begin coding here ...
# Use as many cells as you like

# Load the datasets
df_price = pd.read_csv("data/airbnb_price.csv")
df_room = pd.read_excel("data/airbnb_room_type.xlsx")
df_review = pd.read_csv("data/airbnb_last_review.tsv", sep="\t")

# Quick check of the data
print(df_price.head())
print(df_room.head())
print(df_review.head())
```

```python
# Merge price + room type first
df_merged = df_price.merge(df_room, on="listing_id", how="outer")

# Then merge with review
df_merged = df_merged.merge(df_review, on='listing_id', how='outer')

# Check result 
print(df_merged.shape)
print(df_merged.head())
```

```python
# Convert last_review to datetime
df_merged["last_review"] = pd.to_datetime(df_merged['last_review'],errors='coerce')

# Find earliest and most recent review
first_reviewed = df_merged["last_review"].min()
last_reviewed = df_merged["last_review"].max()

print(first_reviewed)
print(last_reviewed)
```

```python
nb_private_rooms = df_merged["room_type"].str.contains("Private room",case=False,na=False).sum()

print("Number of private rooms:", nb_private_rooms)
```

```python
avg_price = round(pd.to_numeric(df_merged['price'].astype(str).str.replace(" dollars","",case=False),errors="coerce").mean(),2)
print("Average price:", avg_price)
```

```python
review_dates = pd.DataFrame({
    "first_reviewed": [first_reviewed],
    "last_reviewed": [last_reviewed],
    "nb_private_rooms": [nb_private_rooms],
    "avg_price": [avg_price]
})

print(review_dates)
```
