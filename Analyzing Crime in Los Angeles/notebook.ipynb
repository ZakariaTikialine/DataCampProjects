```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", dtype={"TIME OCC": str})
crimes.head()
# Extract hour from TIME OCC (first two characters)
crimes['HOURS OCC'] = crimes['TIME OCC'].str[:2].astype(int)

# Plot frequency of crimes by hour
sns.countplot(data=crimes, x='HOURS OCC')
plt.show()

# Find peak crime hour
peak_crime_hour = crimes['HOURS OCC'].value_counts().idxmax()
print("Peak crime hour:", peak_crime_hour)

# Subset for night crimes (10 PM - 3:59 AM)
subset_night_hours = crimes[(crimes['HOURS OCC'] >= 22) | (crimes['HOURS OCC'] < 4)]

# Count crimes per LOCATION
location_counts = subset_night_hours.groupby("AREA NAME").size()

# Find the location with the highest count
peak_night_crime_location = location_counts.idxmax()

print(peak_night_crime_location)

max_age = crimes["Vict Age"].max()

bins = [0, 17, 25, 34, 44, 54, 64, max_age]
labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

victim_ages = pd.cut(crimes["Vict Age"], bins=bins, labels=labels, include_lowest=True).value_counts().sort_index()
```
